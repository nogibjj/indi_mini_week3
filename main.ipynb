# Soccer Player Data Analysis
This notebook demonstrates the application of functions defined in `lib.py` for analyzing soccer player data. We load a dataset, apply statistical functions, and generate visualizations.


```python
import matplotlib.pyplot as plt
from mylib.lib import (
    load_dataset,
    print_head,
    generate_summary_statistics,
    group_by,
    build_log_histogram,
    create_scatter,
    save_to_markdown,
)

dataset_path = "player_overview.csv"


dataframe = load_dataset(dataset_path)


print("First few rows of the dataset:")
print_head(dataframe)
```

```python
print("\nSummary Statistics:")
summary_stats = generate_summary_statistics(dataframe)
print(summary_stats)


save_to_markdown(summary_stats, "summary_statistics.md")
```

```python
print("\nValue Counts by Position:")
position_counts = group_by(dataframe, "Position")
print(position_counts)
```

```python
build_log_histogram(dataframe, "Goals", "goals_log_histogram.png")
```

```python
create_scatter(dataframe, "Goals", "Assists", "goals_assists_scatter.png")
```

```python
print("\nSaving summary statistics to markdown file:")
save_to_markdown(summary_stats, "player_summary.md")
```
